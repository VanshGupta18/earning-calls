# FinBERT Sandbox Experiment

This notebook validates the `ProsusAI/finbert` model for sentiment analysis on earnings call transcripts.

**Goal:**
1. Load the model from Hugging Face.
2. Run inference on sample earnings call sentences.
3. Extract sentiment labels and probabilities.
4. Measure latency.

```python
import time
import torch
from transformers import AutoTokenizer, AutoModelForSequenceClassification, pipeline
import pandas as pd

model_name = "ProsusAI/finbert"

print(f"Loading {model_name}...")
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForSequenceClassification.from_pretrained(model_name)
nlp = pipeline("sentiment-analysis", model=model, tokenizer=tokenizer)
print("Model loaded successfully.")
```

```python
samples = [
    "We are confident in our growth trajectory for the upcoming fiscal year.",
    "Revenue declined significantly due to macro headwinds and supply chain disruptions.",
    "The operating margin was in line with our previous guidance.",
    "We are seeing unprecedented demand for our new AI-enabled products.",
    "Interest rate volatility remains a key concern for our treasury department."
]

results = []
for text in samples:
    start_time = time.time()
    res = nlp(text)[0]
    latency = (time.time() - start_time) * 1000
    
    results.append({
        "text": text,
        "label": res['label'],
        "score": res['score'],
        "latency_ms": latency
    })

df = pd.DataFrame(results)
display(df)

print(f"\nAverage Latency: {df['latency_ms'].mean():.2f} ms per sentence")
```
